```python
# This cell should be ran to load the functions, but requires no editing.
import os

def give_throw(move_id, throw, number):
    s=''
    t = throw['ThrowKnockbackData']
    s += '''{{{{TrueThrowData-ROA2
| chara = {0}
| moveID = {1}
| throwNumber = Throw{2}
| Damage = {3}
| BaseKnockback = {4}
| KnockbackScaling = {5}
| KnockbackAngle = {6}'''.format(char,
        move_id,
        number,
        t['Damage'],
        t['BaseKnockback'],
        t['KnockbackScaling'],
        t['KnockbackAngle'])
    if(t['HitstunMultiplier'] != 1):
        s+='\n| HitstunMultiplier = {0}'.format(t['HitstunMultiplier'])
    if(not t['bTechable']):
        s+='\n| bTechable = false'
    if(not t['ForceTumble']):
        s+='\n| ForceTumble = false'
    if(t['HitstunAnimationStateOverride'].split('::')[1] != "None"):
        s+='\n| HitstunAnimationStateOverride = {0}'.format(t['HitstunAnimationStateOverride'].split('::')[1])
    s += '\n| ReleaseOffset = {0}'.format(throw['ReleaseOffset'])
    s += '\n}}'
    return s

def give_article(art_id, props):
    s=''
    

    s += '''{{{{ArticleData-ROA2
| chara = {0}
| moveID = {1}
| ArticleName = {2}
'''.format(char,
        art_id,
        props['ArticleName']
)
    for i in ['bIsProjectile', 'bRotateWithVelocity', 'bInheritOwnerChargeValue', 'bIsAttachedToOwner', 'bCanBeHitByOwner', 'bCanDetectOwner']:
        if(i in  props.keys()):
            s += '''| {0} = true\n'''.format(i)
    for i in ['ShouldGetOutOfGroundOnSpawn']:
        if(i in  props.keys()):
            s += '''| {0} = false\n'''.format(i)
    
    for i in ['ParryReaction', 'HasHitReaction', 'GotHitReaction', 'GroundCollisionResponse', 'WallCollisionResponse', 'CeilingCollisionResponse']:
        if(i in props.keys()):
            s += '''| {0} = {1}\n'''.format(i, props[i].split('::')[1])
    s += '}}'
    return s

def give_props(move_id, props):
    s=''
    for hit in props['HitboxOnHitProperties']:
        adv = hit['AdvancedOnHitProperties']
        s += '''{{{{FrameDataNew-ROA2
| chara = {0}
| moveID = {1}
| nameID = {2}
| Damage = {3}
| BaseKnockback = {4}
'''.format(char,
        move_id,
        hit['Name'],
        hit['Damage'],
        hit['BaseKnockback'])
        if hit['KnockbackScaling'] != 0:
            s+= '| KnockbackScaling = {0}\n'.format(hit['KnockbackScaling'])
        s+= '| KnockbackAngle = {0}\n'.format(hit['KnockbackAngle'])
        s+= '''| HitEffectName = {0}
| HitSoundName = {1}\n'''.format(hit['HitEffectName'],
        hit['HitSoundName'])
        if adv['SpecialEffect'].split('::')[1] != 'None':
            s+= '| SpecialEffect = {0}\n'.format(adv['SpecialEffect'].split('::')[1])
        if adv['HitpauseMultiplier'] != 1:
            s+= '| HitpauseMultiplier = {0}\n'.format(adv['HitpauseMultiplier'])
        if adv['ExtraHitpauseForOpponent'] != 0:
            s+= '| ExtraHitpauseForOpponent = {0}\n'.format(adv['ExtraHitpauseForOpponent'])
        if adv['HitpauseMovementType'].split('::')[1] != 'None':
            s+= '| HitpauseMovementType = {0}\n'.format(adv['HitpauseMovementType'].split('::')[1])
        if adv['HitpauseMovementOffsetFromHost']["X"] != 0 or adv['HitpauseMovementOffsetFromHost']["Y"] != 0:
            s+= '| HitpauseMovementOffsetFromHost = {0}\n'.format(adv['HitpauseMovementOffsetFromHost'])
        if adv['HitpauseMovementStrength'] != 0.5:
            s+= '| HitpauseMovementStrength = {0}\n'.format(adv['HitpauseMovementStrength'])
        if adv['SDIMultiplier'] != 1.0:
            s+= '| SDIMultiplier = {0}\n'.format(adv['SDIMultiplier'])
        if adv['ASDIMultiplier'] != -1.0:
            s+= '| ASDIMultiplier = {0}\n'.format(adv['ASDIMultiplier'])
        if not adv['bCanReverse']:
            s+= '| bCanReverse = {0}\n'.format(adv['bCanReverse'])
        if adv['bForceFlinch']:
            s+= '| bForceFlinch = {0}\n'.format(adv['bForceFlinch'])
        if not adv['GroundTechable']:
            s+= '| GroundTechable = {0}\n'.format(adv['GroundTechable'])
        if adv['bIgnoresWeight']:
            s+= '| bIgnoresWeight = {0}\n'.format(adv['bIgnoresWeight'])
        if adv['bAutoFloorhuggable']:
            s+= '| bAutoFloorhuggable = {0}\n'.format(adv['bAutoFloorhuggable'])
        if adv['ProjectileInteraction'].split('::')[1] != "Default":
            s+= '| ProjectileInteraction = {0}\n'.format(adv['ProjectileInteraction'])
        if adv['bForceKnockbackInKnockdown']:
            s+= '| bForceKnockbackInKnockdown = {0}\n'.format(adv['bForceKnockbackInKnockdown'])
        if adv['bPreserveFacing']:
            s+= '| bPreserveFacing = {0}\n'.format(adv['bPreserveFacing'])
        if adv['KnockbackAngleMode'].split('::')[1] != "SpecifiedAngle":
            s+= '| GroundTechable = {0}\n'.format(adv['GroundTechable'])
        if adv['HitstunMultiplier'] != 1:
            s+= '| HitstunMultiplier = {0}\n'.format(adv['HitstunMultiplier'])
        if adv['HitfallHitstunMultiplier'] != 1:
            s+= '| HitfallHitstunMultiplier = {0}\n'.format(adv['HitfallHitstunMultiplier'])
        if adv['ParryReaction'].split('::')[1] != 'Stun':
            s+= '| ParryReaction = {0}\n'.format(adv['ParryReaction'].split('::')[1])
        if adv['GrabPartnerInteraction'].split('::')[1] != 'None':
            s+= '| GrabPartnerInteraction = {0}\n'.format(adv['GrabPartnerInteraction'].split('::')[1])
        if adv['ExtraShieldStun'] != 0:
            s+= '| ExtraShieldStun = {0}\n'.format(adv['ExtraShieldStun'])
        if adv['ShieldDamageMultiplier'] != 1:
            s+= '| ShieldDamageMultiplier = {0}\n'.format(adv['ShieldDamageMultiplier'])
        if adv['ShieldPushbackMultiplier'] != 1:
            s+= '| ShieldPushbackMultiplier = {0}\n'.format(adv['ShieldPushbackMultiplier'])
        if adv['ShieldHitpauseMultiplier'] != 1:
            s+= '| ShieldHitpauseMultiplier = {0}\n'.format(adv['ShieldHitpauseMultiplier'])
        if adv['FullChargeKnockbackMultiplier'] != 1.3:
            s+= '| FullChargeKnockbackMultiplier = {0}\n'.format(adv['FullChargeKnockbackMultiplier'])
        if adv['FullChargeDamageMultiplier'] != 1.6:
            s+= '| FullChargeDamageMultiplier = {0}\n'.format(adv['FullChargeDamageMultiplier'])
        if adv['FinalBaseKnockback'] != 0.0:
            s+= '| FinalBaseKnockback = {0}\n'.format(adv['FinalBaseKnockback'])
        if adv['ForceTumble']:
            s+= '| ForceTumble = {0}\n'.format(adv['ForceTumble'])
        if adv['IgnoreKnockbackArmor']:
            s+= '| IgnoreKnockbackArmor = {0}\n'.format(adv['IgnoreKnockbackArmor'])
        if adv['PreventChaingrabsOnHit']:
            s+= '| PreventChaingrabsOnHit = {0}\n'.format(adv['PreventChaingrabsOnHit'])
        if adv['HitstunAnimationStateOverride'].split('::')[1] != None:
            s+= '| HitstunAnimationStateOverride = {0}\n'.format(adv['HitstunAnimationStateOverride'].split('::')[1])
        s += '}}\n'
    return s

```
