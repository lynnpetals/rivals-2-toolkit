```python

char='Etalus'
# char_data[char] = dict()

stats = dict()

# if('CD_' in char):
filename = 'ART_Eta_IceHalo.json'
f = json.load(open(os.path.join(home_path, char, 'Attacks/Articles', filename)))[0]

def print_move(char, f, filename):
    outputString = ''
    for w in f['Properties']['Windows']:
        outputString += '{{MoveWindows-ROA2 | chara = ' + char + '\n'
        outputString += '| attack = ' + filename.split('_', maxsplit=2)[2].split('.')[0] + '\n'
        outputString += '| window = ' + w['StringTableKey'] + '\n'
        outputString += '| length  = ' + str(w['WindowLengthFrames']) + '\n'
        outputString += '| iasaFrame  = ' + str(w['IasaFrame']) + '\n'
        outputString += '| animStartFrame  = ' + str(w['AnimationStartFrame']) + '\n'
        outputString += '| animLengthFrames  = ' + str(w['AnimationLengthFrames']) + '\n'
        outputString += '| nextWindow = ' + w['NextWindowStringTableKey'] + '\n'


        outputString += '| cancels = '

        for v in w['WindowCancels']:
            outputString += '\n' + v['WindowCancelType'].split('::')[1] + ', from ' + str(v['WindowCancelFrame']) + ' for ' + str(v['WindowCancelLengthFrames']) + ' frames, to ' + v['PostCancelWindowStringTableKey'] + ', requires ' + v['WindowCancelPrerequisite'].split("::")[1] + '. '
            outputString += 'Properties: ['
            if (v['CancellableWhenParried']):
                outputString += 'Cancellable On Parry, '
            if (v['bPreseveWindowTimer']):
                outputString += 'Preserves Window Timer, '
            if (not v['bConsumeCancelInput']):
                outputString += 'Cancel Input Acts Normally, '
            if(v['CancelIntoAttack'].split('::')[1] != ' None'):
                outputString += v['CancelIntoAttack'].split('::')[1]
            outputString += '] \\'

        outputString += '\n| velocityData = '
        for v in w['VelocityData']:
            outputString += '\n(' + v['HorizontalVelocityType'].split('::')[1] + ' [' + str(v['Velocity']['X']) + '], '
            outputString += '' + v['VerticalVelocityType'].split('::')[1] + ' [' + str(v['Velocity']['Y']) + '])'
            outputString += ' from ' + str(v['ApplyVelocityOnWindowFrame']) + ' to ' + str(v['ApplyVelocityUntilFrame'])
            if(len(v['VelocityModifiers']) > 0):
                for m in v['VelocityModifiers']:
                    outputString += '\n> Modifier: ' + (m['ModifierInput'].split('::')[1]) + '/' + m['ModifierVelocityType'].split('::')[1] + ' (' + str(m['ModifierVelocity']['Y']) + ', ' + str(m['ModifierVelocity']['Y']) + ') at ' + str(m['AngleChange']) + 'deg'

            outputString += ' \\'

        outputString += '\n| movementProps = '
        for k, v in w['MovementProperties'].items():
            if(k in ["Gravity", "MaxFallSpeed", "FrictionGround", "FrictionAir", "AirAcceleration", "AirSpeedHorizontalMax"] and v != -1):
                outputString += k + ': ' + str(v) + ' \\ \n'
            if(k == 'CanDrift' and v != "ERivalsAirControlType::Default"):
                outputString += 'Drift: ' + v.split('::')[1] + ' \\ \n'
            if(k == "CanFallThroughPlatforms" and v):
                outputString += 'Can Fall Through Platforms \\ \n'
            if(k == "DiminishingReturnsPercentLostPerUse" and v != 0):
                outputString += 'Diminished Returns Percent Lost Per Use: ' + str(v) + '\\ \n'
        outputString += '\n| hitboxWindows  = \n'
        for v in w['HitboxWindows']:
            for h in f['Properties']['HitboxAttributes']:
                if(h["Name"] == v['HitboxName']):
                    outputString += v['HitboxName'] + ': ' + h['OnHitPropertiesName'] + ' [' + str(v['StartFrame']+1)
                    if(v['LengthInFrames'] == 0 and w['WindowLengthFrames'] > 1):
                        outputString += '-' + str(v['StartFrame']+w['WindowLengthFrames']) + '] \\\n'
                    if(v['LengthInFrames'] == 1 or (v['LengthInFrames'] == 0 and w['WindowLengthFrames'] == 1)):
                        outputString += '] \\\n'
        outputString += '| refreshHasHit = ' + str(w['bRefreshHasHit']) + '\n'
        outputString += '| AdvancedProperties = '
        for k, v in w['AdvancedProperties'].items():
            if(k in ["WindowArmor"] and v!=0):
                outputString += k + ': ' + str(v) + ' \\ \n'
            if(k in ["CanWallJumpOnFrame", "BReverseOnFrame"] and v!=-1):
                outputString += k + ': ' + str(v) + ' \\ \n'
            if(k in ['ToggleECBActiveOnFrame', 'ToggleRootedOnFrame', 'TurnAroundOnFrame', 'ToggleFarFromECBOnFrame', 'ToggleCanFastfallOnFrame', 'ToggleVisibleOnFrame'] and len(v) > 0):
                outputString += k + ': [' 
                for i in v:
                    outputString += str(i) + ', '
                outputString += '] \\ \n'
        outputString += '\n| LedgeProperties  = '

        for k, v in w['LedgeProperties'].items():
            if(k in ["CanGrabLedgeOnFrame", "ReleaseLedgeOnFrame"] and v!=-1):
                outputString += k + ': ' + str(v) + ' \\ \n'
            if(k in ['LedgeGrabBoxOffset', 'LedgeGrabBoxDimensions'] and not (v['X']==0 and v['Y']==0)):
                outputString += k + ': (' + str(v['X']) + ', ' + str(v['Y']) + ') \\ \n'
        
        outputString += '\n}}\n'
    return outputString
def print_article(char, f, filename):
    outputString = ''
    for w in f['Properties']['Windows']:
        outputString += '{{MoveWindows-ROA2 | chara = ' + char + '\n'
        outputString += '| attack = ' + filename.split('_', maxsplit=2)[2].split('.')[0] + '\n'
        outputString += '| window = ' + w['StringTableKey'] + '\n'
        outputString += '| length  = ' + str(w['WindowLengthFrames']) + '\n'
        outputString += '| animStartFrame  = ' + str(w['AnimationStartFrame']) + '\n'
        outputString += '| animLengthFrames  = ' + str(w['AnimationLengthFrames']) + '\n'
        outputString += '| nextWindow = ' + w['NextWindowStringTableKey'] + '\n'
        outputString += '| velocityData = '
        for v in w['VelocityData']:
            outputString += '\n(' + v['HorizontalVelocityType'].split('::')[1] + ' [' + str(v['Velocity']['X']) + '], '
            outputString += '' + v['VerticalVelocityType'].split('::')[1] + ' [' + str(v['Velocity']['Y']) + '])'
            outputString += ' from ' + str(v['ApplyVelocityOnWindowFrame']) + ' to ' + str(v['ApplyVelocityUntilFrame'])
            if(len(v['VelocityModifiers']) > 0):
                for m in v['VelocityModifiers']:
                    outputString += '\n> Modifier: ' + (m['ModifierInput'].split('::')[1]) + '/' + m['ModifierVelocityType'].split('::')[1] + ' (' + str(m['ModifierVelocity']['Y']) + ', ' + str(m['ModifierVelocity']['Y']) + ') at ' + str(m['AngleChange']) + 'deg'

            outputString += ' \\'
        outputString += '\n| hitboxWindows  = \n'
        for v in w['HitboxWindows']:
            outputString += v['HitboxName'] + ' [' + str(v['StartFrame']) + ',' + str(v['LengthInFrames']) + '] \\\n'
        outputString += '| refreshHasHit = ' + str(w['bRefreshHasHit']) + '\n'
        outputString += '| AdvancedProperties = '
        for k, v in w['AdvancedProperties'].items():
            if(k in ["bVisible", "bEcbEnabled", "bArticlePlatformEnabled"] and not v):
                outputString += k + ' \\ \n'
            if(k in ["Gravity", "FrictionGround", "FrictionAir"] and v!=0):
                outputString += k + ': ' + str(v) + ' \\ \n'
            if(k == 'MaxFallSpeed' and v != 9999.0):
                outputString += k + ': ' + str(v) + ' \\ \n'
            if(k in ["bCanFallThroughPlatforms", "bCanBeHitByHitboxesThatCantHitProjectiles", "bCreateDestroyVfxAtEndOfWindow"] and v):
                outputString += 'NOT' + k + ' \\ \n'
        outputString += '\n}}\n'
    return outputString
```
