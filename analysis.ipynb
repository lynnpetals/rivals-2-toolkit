A guide to interpreting the results above:

= Character =
==== MoveName ====
"Window Name": "Frame Duration Start" - "Frame Duration End" ["Window Length"] -> "Next Window"
    "Hit 1": "Hit Duration Start" - "Hit Duration End"
    "Hit 2": "Hit Duration Start" - "Hit Duration End"
    [CANCEL: Trigger>
        IASA: IASA number. If it's +, there is no limit when the action can be cancelled. If there is a duration, it will list the last possible frame that can be cancelled.
        ENDLAG: Endlag number. See above.

For example:

====Ftilt====
Startup 1: 1-12 [12] -> Active 1
Active 1: 13-15 [3] -> Recovery 1
	Pull: 13-15
Recovery 1: 16-37 [22] -> 
	[CANCEL: AttackPressed> 
		IASA: 21-25
		ENDLAG: 5-9
	[CANCEL: AttackPressed> 
		IASA: 21-30
		ENDLAG: 5-14
Startup 2: 1-5 [5] -> Active 2
Active 2: 6-10 [5] -> Recovery 2
	Sour: 6-10
	Sweet: 7-10
Recovery 2: 11-44 [34] -> 

From this, we can get 12 startup, 13-15 active frames, an endlag of 22 and an IASA of 37 + 1 = 38. We can also get the endlag cancels (5-9) on whiff, and (5-14) on shield, with an IASA of 21-25 and 21-30 respectively.
WE also get the second hit, with 5 startup, different active frames depending on hit, and an endlag of 34 with an IASA of 45.
